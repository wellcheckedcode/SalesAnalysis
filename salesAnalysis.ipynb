```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
top_1000_movies=pd.read_csv("imdb_top_1000.csv")
top_5000=pd.read_csv("movies_reviews.csv")

```

```python
movies_financials=top_5000[['original_title','budget','revenue']]
movies_financials.head(10)
```

```python
top_1000_movies=top_1000_movies.rename(columns={'originale_title':'original_title'})

top_1000_movies.to_csv("imdb_top_1000.csv")
top_1000_movies.head(10)
```

```python
movies_financials.columns = movies_financials.columns.str.strip()
top_1000_movies.columns = top_1000_movies.columns.str.strip()
```

```python
print(top_1000_movies.columns)
print(movies_financials.columns)
```

```python
merged=pd.merge(top_1000_movies,movies_financials,how='left',on='original_title')
merged.drop(merged.columns[[0, 1]], axis=1, inplace=True)


merged['revenue']=(merged['revenue']/10000000)
merged['budget']=(merged['budget']/10000000)
merged.head(10)
```

```python
merged20=merged.head(20)
fig, ax = plt.subplots(figsize=(12,8)) 

ax.bar(merged20['original_title'],merged20['budget'],label="budget")
ax.bar(merged20['original_title'],merged20['revenue'],bottom=merged20['budget'],label="revenue")
ax.set_ylabel("Money in Crores")
ax.set_xlabel("Movies Name")
ax.set_xticklabels(merged20['original_title'],rotation=90)
ax2 = ax.twinx()
ax2.plot(merged20['original_title'], merged20['IMDB_Rating'], color='red', marker='o', label="IMDb Rating")
ax2.set_ylabel("IMDb Rating")
handles, labels = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

ax.legend(handles + handles2, labels + labels2,loc="upper right")
plt.tight_layout()
plt.show()
```
